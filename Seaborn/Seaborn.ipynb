```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load Built-in Dataset Example
tips = sns.load_dataset("tips")
print(tips.head())
```

### Univariate Data Visualization

These plots help understand individual feature distributions.

#### Histogram

A histogram visualizes the distribution of a single numerical variable.

```python
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.show()
```

#### KDE Plot

A Kernel Density Estimate (KDE) plot smooths the histogram using a probability density function.

```python
sns.kdeplot(tips['total_bill'], fill=True)
plt.show()
```

#### Box Plot 

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
df = sns.load_dataset('tips')
sns.boxplot(x='smoker', y='total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,hue="day",legend=False)
```

```python
sns.boxplot(data=df,orient='v')
```

```python
# categorize my data based on some other categories

sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

#### Violin Plot

Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot

```python
sns.violinplot(x="total_bill", y="day", data=df,hue="day",legend=False)
```

```python
sns.violinplot(x=tips["total_bill"])
plt.show()
```

#### Dist plot

Dist plot helps us to check the distribution of the columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

### Bivariate and Multivariate Data Visualization

These plots show relationships between two or more variables.

#### Scatter Plot 

Used to visualize the relationship between two numerical variables.

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", size="size", data=tips)
plt.title("Scatterplot of Tips vs Total Bill")
plt.show()
```

#### Pair Plot 

A pairs plot is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

#### Heatmap

A heatmap visualizes correlations between multiple numerical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
tips = sns.load_dataset("tips")

# Select only numeric columns
corr = tips.select_dtypes(include=["number"]).corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

#### Line Plot

Shows trends over time.

```python
sns.lineplot(x="size", y="total_bill", data=tips)
plt.show()
```

### Customizing Seaborn Plots

#### Adding Titles and Labels

```python
sns.histplot(tips['total_bill'])
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

#### Changing Colors

```python
sns.boxplot(x="day", y="total_bill", data=tips, hue="day", legend=False)
plt.show()
```

#### Adjusting Figure Size

```python
plt.figure(figsize=(10,5))
sns.barplot(x="day", y="total_bill", data=tips)
plt.show()
```

### Advanced Seaborn Concepts for ML

#### FacetGrid

Creates multiple plots based on categorical variables.

```python
g = sns.FacetGrid(tips, col="sex")
g.map(sns.histplot, "total_bill")
plt.show()
```

#### Regression Plot

Adds a regression line to a scatter plot.

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.show()
```

#### Joint Plot 

A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

#### Swarm Plot

Similar to a violin plot but shows individual points.

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.show()
```

#### Feature Relationships for Model Selection
