### Types of Plots in Matplotlib

#### Line Plot

Connects data points with a line, useful for trends (e.g., loss curves in ML).

 - Simple Line Graph

```python
plt.plot([1, 2, 3], [1, 4, 9])
plt.show()
```

 - Multiple Lines

```python
plt.plot([1, 2, 3], [1, 4, 9], label='x^2')
plt.plot([1, 2, 3], [1, 2, 3], label='x')
plt.legend()
plt.show()
```

 - Line Styles

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

plt.plot(x, y, '-')     # Solid
plt.plot(x, [y[i]-2 for i in range(4)], '--')  # Dashed
plt.plot(x, [y[i]-4 for i in range(4)], ':')   # Dotted
plt.plot(x, [y[i]-6 for i in range(4)], '-.')  # Dash-dot
plt.show()
```

 - Colors

```python
plt.plot(x, y, 'b-')         # Blue solid
plt.plot(x, [y[i]-2 for i in range(4)], 'r--')  # Red dashed
plt.plot(x, [y[i]-4 for i in range(4)], 'g:')   # Green dotted
plt.plot(x, [y[i] - 6 for i in range(4)], color='#FF5733', linestyle='-.')   # Orange dash-dot
plt.show()
```

 - Markers

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]

# Plot each line with a different marker
plt.plot(x, y, '-o', label='Circle (o)')          # Circle
plt.plot(x, [i - 2 for i in y], '-s', label='Square (s)')       # Square
plt.plot(x, [i - 4 for i in y], '-^', label='Triangle Up (^)')  # Triangle (up)
plt.plot(x, [i - 6 for i in y], '-v', label='Triangle Down (v)') # Triangle (down)
plt.plot(x, [i - 8 for i in y], '-x', label='X (x)')            # X
plt.plot(x, [i - 10 for i in y], '-+', label='Plus (+)')        # Plus
plt.plot(x, [i - 12 for i in y], '-*', label='Star (*)')        # Star
plt.plot(x, [i - 14 for i in y], '-.', label='Point (.)')       # Point
plt.plot(x, [i - 16 for i in y], '-d', label='Diamond (d)')     # Diamond

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('All Common Markers in Matplotlib')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))  # Legend outside plot
plt.grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plot
plt.show()
```

 - Annotations

```python
plt.plot([1, 2, 3], [1, 4, 9])
plt.annotate('Peak', xy=(3, 9), xytext=(2, 10), arrowprops=dict(facecolor='black'))
plt.show()
```

#### Scatter Plot

Shows individual data points, great for visualizing clusters or predictions.

 - Simple Scatter

```python
plt.scatter([1, 2, 3], [2, 4, 6])
plt.show()
```

 - Marker Styles

```python
plt.scatter([1, 2, 3], [2, 4, 6], marker='s')  # Square markers
plt.show()
```

 - Colormaps

```python
plt.scatter([1, 2, 3], [2, 4, 6], c=[0, 1, 2], cmap='viridis')
plt.colorbar()
plt.show()
```

 - Transparency

```python
plt.scatter([1, 2, 3], [2, 4, 6], alpha=0.5)  # Semi-transparent
plt.show()
```

#### Bar Chart

Displays data with bars, useful for comparing quantities (e.g., feature importance).

 - Vertical Bar

```python
plt.bar(['A', 'B', 'C'], [3, 7, 2])
plt.show()
```

 - Horizontal Bar

```python
plt.barh(['A', 'B', 'C'], [3, 7, 2])
plt.show()
```

 - Stacked Bar

```python
plt.bar([1, 2, 3], [2, 3, 4])
plt.bar([1, 2, 3], [1, 2, 1], bottom=[2, 3, 4])
plt.show()
```

#### Histogram

Shows the distribution of data, common for feature analysis in ML.

 - Simple Histogram

```python
data = [1, 1, 2, 3, 3, 3, 4]
plt.hist(data, bins=4)
plt.show()
```

 - Density Plot

```python
plt.hist(data, bins=4, density=True)
plt.show()
```

 - Overlay

```python
plt.hist([1, 1, 2], bins=3, alpha=0.5)
plt.hist([2, 3, 3], bins=3, alpha=0.5)
plt.show()
```

#### Box Plot

Summarizes data spread and outliers, useful for comparing distributions.

 - Simple Boxplot


```python
plt.boxplot([1, 2, 3, 10])
plt.show()
```

 - Multiple Boxplots

```python
plt.boxplot([[1, 2, 3], [4, 5, 10]])
plt.show()
```

#### Heatmap



Visualizes a matrix with colors, often for correlations in ML.

```python
import numpy as np
data = np.array([[1, 0.8], [0.8, 1]])
plt.imshow(data, cmap='coolwarm')
plt.colorbar()
plt.show()
```

### Customizing Plots

#### Labels & Titles

Add descriptions to axes and the plot.

```python
plt.plot([1, 2], [1, 4])
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)
plt.title('My Plot', fontsize=16)
plt.show()
```

#### Legends

Explain what lines or points represent.

```python
plt.plot([1, 2], [1, 4], label='Line 1')
plt.legend(loc='upper left')
plt.show()
```

#### Grid & Spines

Add a grid or adjust plot borders.

```python
plt.plot([1, 2], [1, 4])
plt.grid(True)
plt.gca().spines['top'].set_visible(False)  # Hide top spine
plt.show()
```

#### Ticks & Tick Labels

Customize axis markers

```python
plt.plot([1, 2], [1, 4])
plt.xticks([1, 2], ['Start', 'End'])
plt.show()
```

#### Text & Annotations

Add text or arrows to highlight points.

```python
plt.plot([1, 2], [1, 4])
plt.text(1.5, 2, 'Middle')
plt.show()
```

### Subplots & Multiple Plots

#### Simple Subplots

Create multiple plots in one figure using subplot

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2], [1, 4])
plt.subplot(1, 2, 2)
plt.plot([1, 2], [4, 1])
plt.show()
```

#### Using plt.subplots()

Cleaner way to create subplots.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot([1, 2], [1, 4])
ax2.plot([1, 2], [4, 1])
plt.show()
```

#### Sharing Axes

Sync axes across subplots.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot([1, 2], [1, 4])
ax2.plot([1, 2], [2, 3])
plt.show()
```

#### Twin Axes

Use two Y-axes for different scales (e.g., loss vs. accuracy).


```python
fig, ax1 = plt.subplots()
ax1.plot([1, 2], [1, 4], 'b-')
ax2 = ax1.twinx()
ax2.plot([1, 2], [10, 20], 'r-')
plt.show()
```

### Advanced Customization

#### Controlling Figure Size & Resolution

Adjust the size and quality of the plot.

```python
plt.figure(figsize=(8, 4), dpi=100)
plt.plot([1, 2], [1, 4])
plt.show()
```

#### Adjusting Aspect Ratio



Control the shape of the plot.

```python
plt.plot([1, 2], [1, 2])
plt.axis('equal')
plt.show()
```

#### Adding Images

Display an image within a plot.

```python
img = np.random.rand(10, 10)
plt.imshow(img, cmap='gray')
plt.show()
```

### 3D Plotting



Create 3D visualizations for complex data.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter([1, 2], [1, 2], [1, 3])
plt.show()
```

### Interactive Plots

Add sliders or buttons for interactivity.

```python
from matplotlib.widgets import Slider
fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)
line, = ax.plot([1, 2], [1, 4])
slider_ax = plt.axes([0.25, 0.1, 0.65, 0.03])
slider = Slider(slider_ax, 'Slope', 1, 10, valinit=4)
def update(val):
    line.set_ydata([1, slider.val])
    fig.canvas.draw_idle()
slider.on_changed(update)
plt.show()
```
